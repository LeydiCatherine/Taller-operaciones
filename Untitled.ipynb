```python
df_pitt = df[df["city"] == "Pittsburgh"]
```

```python
df_pitt.head()
```

```python
df_vegas = df[df["city"] == "Las Vegas"]
```

```python
pitt_stars = df_pitt["stars"]
vegas_stars = df_vegas["stars"]
```

```python
plt.hist(
    pitt_stars,
    alpha = 0.3,
    color = 'yellow',
    label = 'Pittsburgh',
    bins = 'auto'
    
)

plt.hist(
    vegas_stars,
    alpha = 0.3,
    color = 'red',
    label = 'Las Vegas',
    bins = 'auto'
    
)

plt.xlabel('Rating')
plt.ylabel('Number of Rating Score')

plt.legend(loc = 'best')

plt.title('Review distribution of Pittsburgh and Las Vegas')

plt.show()
```

```python
plt.hist(
    [pitt_stars, vegas_stars],
    alpha = 0.7,
    color = ['red', 'blue'],
    label = ['Pittsburgh', 'Las Vegas'],
    bins = 'auto'            
)

plt.xlabel('Rating')
plt.ylabel('Number of Rating Score')

plt.legend(loc = 'best')

plt.title('Review distribution of Pittsburgh and Las Vegas')

plt.show()
```
